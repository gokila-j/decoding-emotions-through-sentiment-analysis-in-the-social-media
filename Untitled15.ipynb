```python

!pip install gradio transformers torch

import gradio as gr
from transformers import pipeline

# Emotion-to-emoji mapping
emotion_emoji_map = {
    "joy": "😄", "sadness": "😢", "anger": "😠", "fear": "😱", "disgust": "🤢",
    "surprise": "😲", "love": "❤️", "admiration": "👏", "amusement": "😂",
    "gratitude": "🙏", "nervousness": "😬", "optimism": "🌟", "pride": "🏅",
    "disappointment": "😞", "relief": "😌", "caring": "🤗", "remorse": "😔",
    "confusion": "😕", "realization": "💡", "neutral": "😐"
}

# Load emotion classifier
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base")

# Define function for Gradio
def analyze_emotion(text):
    result = classifier(text)[0]
    emotion = result['label']
    score = result['score']
    emoji_icon = emotion_emoji_map.get(emotion.lower(), "❓")
    return f"Detected Emotion: {emotion} {emoji_icon} (Confidence: {score:.2f})"

# Gradio interface
gr.Interface(
    fn=analyze_emotion,
    inputs=gr.Textbox(lines=2, placeholder="Enter a social media message..."),
    outputs="text",
    title="Emotion-to-Emoji Classifier",
    description="Detects the emotion of a message and displays the corresponding emoji."
).launch()
```
